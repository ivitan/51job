# 薪酬统一处理(K/月)

```python
import re
low_salary=[]
high_salary=[]
avg_salary = []

f_csv = df['salary']
for row in f_csv:
    if row == '暂无信息':
        low_salary.append(0)
        high_salary.append(0)
        avg_salary.append(0)
        
    elif "千/月" in row:
        low = float(re.search((r'(\d+\.?\d*)-(\d+\.?\d*)'),row).group(1))
        high = float(re.search((r'(\d+\.?\d*)-(\d+\.?\d*)'),row).group(2))
        low_salary.append(low)
        high_salary.append(high)
        avg_salary.append((low+high)/2)
        
    elif "万/月" in row:
        low = float(re.search((r'(\d+\.?\d*)-(\d+\.?\d*)'),row).group(1))
        high = float(re.search((r'(\d+\.?\d*)-(\d+\.?\d*)'),row).group(2))
        low_salary.append(low * 10)
        high_salary.append(high * 10)
        avg_salary.append((low+high)*10/2)
        
    elif "万/年" in row:
        low = float(re.search((r'(\d+\.?\d*)-(\d+\.?\d*)'),row).group(1))
        high = float(re.search((r'(\d+\.?\d*)-(\d+\.?\d*)'),row).group(2))
        low_salary.append(low / 12 * 10)
        high_salary.append(high / 12 * 10)
        avg_salary.append((low+high)*10/2)
        
    elif "/天" in row:
        low = float(re.search((r'(\d+\.?\d*).*'),row).group(1))
        low_salary.append(low * 21 / 10 )
        high_salary.append(high * 30 / 10)
        avg_salary.append((low+high)/2)
        
    elif "千以下/月" in row:
        low = float(re.search((r'(\d+\.?\d*).*'),row).group(1))
        low_salary.append(0)
        high_salary.append(low)
        avg_salary.append((low+high)/2)  
    
    elif "万以下/年" in row:
        low = float(re.search((r'(\d+\.?\d*).*'),row).group(1))
        low_salary.append(0)
        high_salary.append(low/12*10)
        avg_salary.append(low/12*10/2)  
    
    elif "万以下/月" in row:
        low = float(re.search((r'(\d+\.?\d*).*'),row).group(1))
        low_salary.append(0)
        high_salary.append(low)
        avg_salary.append((low+high)/2) 
        
df['low_salary'],df['high_salary'],df['avg_salary'] = low_salary,high_salary,avg_salary
del df['salary']
df
```

# 词频统计

```python
import jieba
import matplotlib.pyplot as plt
from  wordcloud import WordCloud

content = df['jpb_info']

# 读取停用词
with open('stopwords.txt', 'r',encoding='utf-8') as fp:
        stopwords = fp.read().split('\n')

contents = ''.join([txt for txt in content if txt not in stopwords and txt !='xa0'])
wordList = jieba.cut(contents)

# 词频统计
data = {}
for word in wordList:
    if len(word)==1:
        continue
    else:data[word]=data.get(word,0)+1

# 匿名函数获取每个单词的次数项统，降序
result = list(data.items())
result.sort(key=lambda x:x[1],reverse=True) # 排序
for i in range(20):
    print(result[i])
    
# 设置词云
wordSplit = " ".join(dict(result))
# backgroud_Image = plt.imread('ivitan.png')
wc = WordCloud(
            background_color='white',# 设置背景颜色
            # mask=backgroud_Image,# 设置背景图片
            max_words=100,  # 设置最大显示的字数
            stopwords = stopwords, #设置停用词
            font_path="/usr/share/fonts/SimHei.ttf",
            # 设置中文字体，使得词云可以显示（词云默认字体是“DroidSansMono.ttf字体库”，不支持中文）
            max_font_size=50,  # 设置字体最大值
            random_state=30,  # 设置有多少种随机生成状态，即有多少种配色方案
            )
mywc = wc.generate(wordSplit)  # 生成词云

# 展示词云图
plt.imshow(mywc)
plt.axis("off")
wc.to_file('images/WordCloud.png') 
plt.show()
```

# 数据分析岗位中各类型企业所占比例

```python
import matplotlib
import matplotlib.pyplot as plt
CompanyNature_Count = df['com_nature'].value_counts()
#设置中文字体
font = {'family': 'SimHei'}
matplotlib.rc('font', **font)
fig = plt.figure(figsize = (8, 8))
#绘制饼图, 参数pctdistance表示饼图内部字体离中心距离, labeldistance则是label的距离, radius指饼图的半径
patches, l_text, p_text = plt.pie(
    CompanyNature_Count, 
    autopct = '%.2f%%',
    pctdistance = 0.6,
    labels = CompanyNature_Count.index,
    labeldistance=1.1, 
    radius = 1,
    textprops = {'fontsize':15,'color':'w'},
    wedgeprops = {'linewidth': 1.5, 'edgecolor':'black'},# 设置饼图内外边界的属性值
)
m , n= 0.02, 0.028
for t in l_text[7: 11]:
    t.set_y(m)
    m += 0.1
for p in p_text[7: 11]:
    p.set_y(n)
    n += 0.1
plt.legend(bbox_to_anchor=(0.9, 0.5)) 
plt.title('岗位中各类型企业所占比例', fontsize=24)
plt.savefig('images/岗位中各类型企业所占比例.png')
```
